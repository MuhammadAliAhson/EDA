```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel("/kaggle/input/demensia/oasis_longitudinal_demographics.xlsx")
print(df.head())
```

```python
sns.countplot(x='Group',data=df)
plt.show()
```

```python
sns.scatterplot(x='MMSE',y='Age',data=df,hue='Group')
plt.show()
```

```python
sns.countplot(x="SES",data=df,hue="Group")
plt.show()
```

```python
sns.distplot(df['SES'])
plt.show()
```

```python
sns.distplot(df["MMSE"],hist=False)
plt.show()
```

```python
#print(sns.get_dataset_names())

#gf= sns.load_dataset('car_crashes')
#print(gf.head())
sns.jointplot(x=df["Age"],y=df['MMSE'],data=df,hue=df['Group'])
plt.show()
```

```python
sns.jointplot(x=df['Age'],y=df['CDR'],data=df,kind='hex')
plt.show()
```

```python
sns.kdeplot(df['ASF'])
plt.show()
```

```python
sns.pairplot(df,hue='Group',palette="Set1")
plt.show()
```

```python
sns.barplot(data=df,x='M/F',y='Age',estimator='std')
plt.show()
```

```python
sns.boxplot(x='M/F',y='MMSE',data=df,hue='Group')
plt.show()
```

```python
sns.violinplot(x='CDR',y='ASF',data=df,hue='Group')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x='M/F',y='nWBV',data=df,hue='Group',jitter=True,dodge=True,palette="Set1")
plt.show()
```
